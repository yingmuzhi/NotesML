将特定目录下的全部文件的后缀增加.csv后缀

```python
'''增加后缀并复制到特定目录'''
import os
import shutil

source_folder = '/home/yingmuzhi/NotesML/src/origin/20240312'  # 原始文件夹路径
destination_folder = '/home/yingmuzhi/NotesML/src/csv'  # 目标文件夹路径

# 遍历原始文件夹下的所有文件
for filename in os.listdir(source_folder):
    # 检查文件是否没有后缀名
    if '.' not in filename:
        # 构造新的文件名，添加 '.csv' 后缀
        new_filename = filename + '.csv'
        # 构造源文件和目标文件的路径
        source_file = os.path.join(source_folder, filename)
        destination_file = os.path.join(destination_folder, new_filename)
        # 使用 shutil.copyfile() 方法复制文件到目标文件夹
        shutil.copyfile(source_file, destination_file)
```

遍历文件名

```python
import os

directory = '/home/yingmuzhi/NotesML/src/csv'  # 目录路径

# 初始化一个空列表，用于存储文件名
file_names = []

# 遍历目录下的所有文件
for filename in os.listdir(directory):
    # 检查文件是否以 '.csv' 结尾
    if filename.endswith('.csv'):
        # 将文件名中的 '.csv' 后缀删除，并存储在列表中
        file_names.append(filename[:-4])  # 删除后缀 '.csv' 的部分

# 打印结果列表
print(file_names)
```

将文件中的元素转为竖着排放，

```python
import convert_data

input_path = "/home/yingmuzhi/NotesML/src/csv"
save_path = "/home/yingmuzhi/NotesML/src/temp_csv"
for file_name in file_names:
    convert_data.convert_file(input_path=input_path, file_name=file_name, save_path=save_path)
```
